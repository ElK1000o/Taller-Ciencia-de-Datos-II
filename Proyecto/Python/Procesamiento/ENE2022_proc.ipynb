```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../../input/data/ENE2022.csv', low_memory=False, sep=';')
data
```

```python
# Copiamos df original en variable para procesar
datos_proc = data.copy()
```

```python
# Exploración de variables I
print(data.shape, "\n\n")
print(data.columns, "\n\n")
print(data.dtypes, "\n\n")
print(data['sexo'].value_counts(), "\n\n")
print(data['region'].value_counts(), "\n\n")
print(data['edad'].value_counts(), "\n\n")
print(data['edad'].describe(), "\n\n")
```

```python
# Exploración de variables II
print(data['b1'].value_counts().sort_index(), '\n\n')
print(data['b13_rev4cl_caenes'].value_counts().sort_index(), '\n\n')
print(data['b14_rev4cl_caenes'].value_counts().sort_index(), '\n\n')
print(data['r_p_rev4cl_caenes'].value_counts().sort_index(), '\n\n')
```

```python
# Diccionarios de Recodificación
ciiu4 = {
    1: 'A Agricultura, ganadería, silvicultura y pesca',
    2: 'B Explotación de minas y canteras',
    3: 'C Industrias manufactureras',
    4: 'D Suministro de electricidad, gas, vapor y aire acondicionado',
    5: 'E Suministro de agua; evacuación de aguas residuales, gestión de desechos y descontaminación',
    6: 'F Construcción',
    7: 'G Comercio al por mayor y al por menor; reparación de vehiculos automotores y motocicletas',
    8: 'H Transporte y almacenamiento',
    9: 'I Actividades de alojamiento y de servicio de comidas',
    10: 'J Información y comunicaciones',
    11: 'K Actividades financieras y de seguros',
    12: 'L Actividades inmobiliarias',
    13: 'M Actividades profesionales, científicas y tecnicas',
    14: 'N Actividades de servicios administrativos y de apoyo',
    15: 'O Administración pública y defensa; planes de seguridad social de afiliación obligatoria',
    16: 'P Enseñanza',
    17: 'Q Actividades de atención de la salud humana y de asistencia social',
    18: 'R Actividades artísticas, de entretenimiento y recreativas',
    19: 'S Otras actividades de servicios',
    20: 'T Actividades de los hogares como empleadores; actividades no diferenciadas de los hogares como productores de bienes y servicios para uso propio',
    21: 'U Actividades de organizaciones y órganos extraterritoriales',
}

reemplazos = {
    'b1' : {999: np.nan},
    'b13_rev4cl_caenes': {999: np.nan},
    'b14_rev4cl_caenes': {999: np.nan},
    'r_p_rev4cl_caenes': {999: np.nan},
    'activ': {2: 0,
              3: 0}
}
```

```python
for col, repl in reemplazos.items():
    datos_proc[col] = datos_proc[col].replace(repl)
```

```python
# Procesamiento
datos_proc = (
    datos_proc
    .assign(
        sexo_str=lambda x: x['sexo'].map({1: "Hombre", 2: "Mujer"}),

        sexo=lambda x: x['sexo'].map({1: 0, 2: 1}),

        edad_tr=lambda x: np.select(
            [x['edad'].between(18, 39), 
             x['edad'].between(40, 64), 
             x['edad'] >= 65],
            ["Jovenes", "Adultos", "Adultos mayores"],
            default="Desconocido" 
        ),

        sexo_edad=lambda x: np.select(
            [(x['sexo_str'] == "Mujer") & (x['edad_tr'] == "Jovenes"),
             (x['sexo_str'] == "Mujer") & (x['edad_tr'] == "Adultos"),
             (x['sexo_str'] == "Mujer") & (x['edad_tr'] == "Adultos mayores"),
             (x['sexo_str'] == "Hombre") & (x['edad_tr'] == "Jovenes"),
             (x['sexo_str'] == "Hombre") & (x['edad_tr'] == "Adultos"),
             (x['sexo_str'] == "Hombre") & (x['edad_tr'] == "Adultos mayores")],
            ["Mujer joven", "Mujer adulta", "Mujer mayor", "Hombre joven", "Hombre adulto", "Hombre mayor"],
            default="Desconocido"
        ),
        
        region_str=lambda x: x['region'].map({
            1: "Región de Tarapacá",
            2: "Región de Antofagasta",
            3: "Región de Atacama",
            4: "Región de Coquimbo",
            5: "Región de Valparaíso",
            6: "Región de O'Higgins",
            7: "Región del Maule",
            8: "Región del Biobío",
            9: "Región de La Araucanía",
            10: "Región de Los Lagos",
            11: "Región de Aysén",
            12: "Región de Magallanes y Antártica",
            13: "Región Metropolitana",
            14: "Región de Los Ríos",
            15: "Región de Arica y Parinacota",
            16: "Región de Ñuble"
        }),
        
       educa_str=lambda x: np.select(
            [x['nivel'].notna() & x['nivel'].isin([0, 1, 2, 88, 99, 902]),
            x['nivel'].notna() & (x['nivel'] == 3),
            x['nivel'].notna() & x['nivel'].isin([4, 6]),
            x['nivel'].notna() & x['nivel'].isin([5, 7, 8]) & (x['termino_nivel'] == 1),
            x['nivel'].notna() & x['nivel'].isin([5, 7, 8]) & (x['termino_nivel'] == 2),
            x['nivel'].notna() & (x['nivel'] == 9) & (x['termino_nivel'] == 1),
            x['nivel'].notna() & (x['nivel'] == 9) & (x['termino_nivel'] == 2),
            x['nivel'].notna() & x['nivel'].isin([10, 11, 12]) & (x['termino_nivel'] == 1),
            x['nivel'].notna() & x['nivel'].isin([10, 11, 12]) & (x['termino_nivel'] == 2)],
            [np.nan, "Educación Básica", "Educación Media",
            "Técnica completa", "Técnica incompleta", "Universitaria completa", 
            "Universitaria incompleta", "Postgrado", "Universitaria completa"],
            default=np.nan
        ),
        
       educa=lambda x: np.select(
            [x['nivel'].notna() & x['nivel'].isin([0, 1, 2, 88, 99, 902]),
            x['nivel'].notna() & (x['nivel'] == 3),
            x['nivel'].notna() & x['nivel'].isin([4, 6]),
            x['nivel'].notna() & x['nivel'].isin([5, 7, 8]) & (x['termino_nivel'] == 1),
            x['nivel'].notna() & x['nivel'].isin([5, 7, 8]) & (x['termino_nivel'] == 2),
            x['nivel'].notna() & (x['nivel'] == 9) & (x['termino_nivel'] == 1),
            x['nivel'].notna() & (x['nivel'] == 9) & (x['termino_nivel'] == 2),
            x['nivel'].notna() & x['nivel'].isin([10, 11, 12]) & (x['termino_nivel'] == 1),
            x['nivel'].notna() & x['nivel'].isin([10, 11, 12]) & (x['termino_nivel'] == 2)],
            [np.nan, 1, 2, 4, 3, 6, 5, 7, 6],
            default=np.nan
        ),

        grupoOcup_ciuo08_str=lambda x: x['b1'].map({
            1: 'Directores y gerentes',
            2: 'Profesionales científicos e intelectuales',
            3: 'Técnicos y profesionales de nivel medio',
            4: 'Personal de apoyo administrativo',
            5: 'Trabajadores de los servicios y vendedores de comercios y mercados',
            6: 'Agricultores y trabajadores calificados agropecuarios, forestales y pesqueros',
            7: 'Oficiales, operarios y artesanos de artes mecánicas y de otros oficios',
            8: 'Operadores de instalaciones y máquinas y ensambladores',
            9: 'Ocupaciones elementales',
            10: 'Otros no identificados',
        }),
        
        categoria_ocupacion_str=lambda x: x['categoria_ocupacion'].map({
            0: 'No corresponde',
            1: 'Empleador',
            2: 'Cuenta propia (independiente)',
            3: 'Asalariado sector privado',
            4: 'Asalariado sector público',
            5: 'Personal de servicio doméstico puertas afuera',
            6: 'Personal de servicio doméstico puertas adentro',
        }),
        
        ocup_form_str=lambda x: x['ocup_form'].map({
            1: 'Ocupado formal',
            2: 'Ocupado informal',
        }),
        
        proveedor_str=lambda x: x['proveedor'].map({
            0: 'No proveedor',
            1: 'Proveedor principal',
        }),
        
        parentesco_str=lambda x: x['parentesco'].map({
            1: 'Persona proveedora principal',
            2: 'Cónyuge',
            3: 'Conviviente',
            4: 'Hijo/a o hijastro/a',
            5: 'Yerno o nuera',
            6: 'Nieto/a',
        }),
        
        empresaPaga_ciiu4_str=lambda x: x['b13_rev4cl_caenes'].map(ciiu4),

        empresaTrabaja_ciiu4_str=lambda x: x['b14_rev4cl_caenes'].map(ciiu4),
        
        empresaDueno_ciiu4_str=lambda x: x['r_p_rev4cl_caenes'].map(ciiu4)
        )
        
    .rename(columns={'idrph': 'id_persona'}) # Renombrar columnas (en caso que corresponda)
    .drop_duplicates(subset=['id_persona']) # Eliminar duplicados en columnas (en caso que corresponda)
    .dropna(subset=['id_persona']) # Eliminar NaN de columnas (en caso que corresponda)
    .query('edad >= 18') # Mantener rango de columnas (en caso que corresponda)
)
```

```python
datos_proc = datos_proc[['sexo', 
                         'sexo_str', 
                         'edad', 
                         'edad_tr', 
                         'sexo_edad', 
                         'region', 
                         'region_str', 
                         'educa',
                         'educa_str',
                         'habituales',
                         'efectivas',
                         'b1',
                         'grupoOcup_ciuo08_str',
                         'activ',
                         'categoria_ocupacion',
                         'categoria_ocupacion_str',
                         'ocup_form',
                         'ocup_form_str',
                         'proveedor', 
                         'proveedor_str', 
                         'parentesco',
                         'parentesco_str',
                         'empresaPaga_ciiu4_str',
                         'empresaTrabaja_ciiu4_str',
                         'empresaDueno_ciiu4_str'
                         ] + 
                        [col for col in datos_proc.columns if 'rev4cl' in col] +
                        [col for col in datos_proc.columns if col.startswith('c2_1')]]

datos_proc.rename(columns={
    'c2_1_1': 'hrs_diarias',
    'c2_1_2': 'dias_sem',
    'c2_1_3': 'hrs_sem',
    'b1': 'grupoOcup_ciuo08',
    'b13_rev4cl_caenes': 'empresaPaga_ciiu4',
    'b14_rev4cl_caenes': 'empresaTrabaja_ciiu4',
    'r_p_rev4cl_caenes': 'empresaDueno_ciiu4',
}, inplace=True)
```

```python
# Revisión de variables I
print(datos_proc.shape, "\n\n")
print(datos_proc.columns, "\n\n")
print(datos_proc.dtypes, "\n\n")
print(datos_proc['sexo'].value_counts(), '\n\n')
print(datos_proc['edad_tr'].value_counts(), '\n\n')
print(datos_proc['sexo_edad'].value_counts(), '\n\n')
print(datos_proc['region'].value_counts(), '\n\n')
print(datos_proc['educa'].value_counts(), '\n\n')
```

```python
# Revisión de variables II
print(datos_proc['grupoOcup_ciuo08_str'].value_counts().sort_index(), '\n\n')
print(datos_proc['empresaPaga_ciiu4_str'].value_counts().sort_index(), '\n\n')
print(datos_proc['empresaTrabaja_ciiu4_str'].value_counts().sort_index(), '\n\n')
print(datos_proc['empresaDueno_ciiu4_str'].value_counts().sort_index(), '\n\n')
```

```python
# Guardar datos
datos_proc.to_csv("../../output/data/ENE2022_proc.csv", index=False)
```
