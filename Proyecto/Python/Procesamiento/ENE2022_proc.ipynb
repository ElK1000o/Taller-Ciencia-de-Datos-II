```python
# Copiamos df original en variable para procesar
datos_proc = data.copy()
```

```python
# Diccionario de RecodificaciÃ³n
ciiu4 = {
    1: 'A Agricultura, ganaderÃ­a, silvicultura y pesca',
    2: 'B ExplotaciÃ³n de minas y canteras',
    3: 'C Industrias manufactureras',
    4: 'D Suministro de electricidad, gas, vapor y aire acondicionado',
    5: 'E Suministro de agua; evacuaciÃ³n de aguas residuales, gestiÃ³n de desechos y descontaminaciÃ³n',
    6: 'F ConstrucciÃ³n',
    7: 'G Comercio al por mayor y al por menor; reparaciÃ³n de vehiculos automotores y motocicletas',
    8: 'H Transporte y almacenamiento',
    9: 'I Actividades de alojamiento y de servicio de comidas',
    10: 'J InformaciÃ³n y comunicaciones',
    11: 'K Actividades financieras y de seguros',
    12: 'L Actividades inmobiliarias',
    13: 'M Actividades profesionales, cientÃ­ficas y tecnicas',
    14: 'N Actividades de servicios administrativos y de apoyo',
    15: 'O AdministraciÃ³n pÃºblica y defensa; planes de seguridad social de afiliaciÃ³n obligatoria',
    16: 'P EnseÃ±anza',
    17: 'Q Actividades de atenciÃ³n de la salud humana y de asistencia social',
    18: 'R Actividades artÃ­sticas, de entretenimiento y recreativas',
    19: 'S Otras actividades de servicios',
    20: 'T Actividades de los hogares como empleadores; actividades no diferenciadas de los hogares como productores de bienes y servicios para uso propio',
    21: 'U Actividades de organizaciones y Ã³rganos extraterritoriales',
    999: '',
}
```

```python
# Procesamiento
datos_proc = (
    datos_proc
    .assign(
        sexo=lambda x: x['sexo'].map({1: "Hombre", 2: "Mujer"}),

        edad_tr=lambda x: np.select(
            [x['edad'].between(18, 39), 
             x['edad'].between(40, 64), 
             x['edad'] >= 65],
            ["Jovenes", "Adultos", "Adultos mayores"],
            default="Desconocido" 
        ),

        sexo_edad=lambda x: np.select(
            [(x['sexo'] == "Mujer") & (x['edad_tr'] == "Jovenes"),
             (x['sexo'] == "Mujer") & (x['edad_tr'] == "Adultos"),
             (x['sexo'] == "Mujer") & (x['edad_tr'] == "Adultos mayores"),
             (x['sexo'] == "Hombre") & (x['edad_tr'] == "Jovenes"),
             (x['sexo'] == "Hombre") & (x['edad_tr'] == "Adultos"),
             (x['sexo'] == "Hombre") & (x['edad_tr'] == "Adultos mayores")],
            ["Mujer joven", "Mujer adulta", "Mujer mayor", "Hombre joven", "Hombre adulto", "Hombre mayor"],
            default="Desconocido"
        ),
        
        region=lambda x: x['region'].map({
            1: "RegiÃ³n de TarapacÃ¡",
            2: "RegiÃ³n de Antofagasta",
            3: "RegiÃ³n de Atacama",
            4: "RegiÃ³n de Coquimbo",
            5: "RegiÃ³n de ValparaÃ­so",
            6: "RegiÃ³n de O'Higgins",
            7: "RegiÃ³n del Maule",
            8: "RegiÃ³n del BiobÃ­o",
            9: "RegiÃ³n de La AraucanÃ­a",
            10: "RegiÃ³n de Los Lagos",
            11: "RegiÃ³n de AysÃ©n",
            12: "RegiÃ³n de Magallanes y AntÃ¡rtica",
            13: "RegiÃ³n Metropolitana",
            14: "RegiÃ³n de Los RÃ­os",
            15: "RegiÃ³n de Arica y Parinacota",
            16: "RegiÃ³n de Ã‘uble"
        }),

        b1=lambda x: x['b1'].map({
            1: 'Directores y gerentes',
            2: 'Profesionales cientÃ­ficos e intelectuales',
            3: 'TÃ©cnicos y profesionales de nivel medio',
            4: 'Personal de apoyo administrativo',
            5: 'Trabajadores de los servicios y vendedores de comercios y mercados',
            6: 'Agricultores y trabajadores calificados agropecuarios, forestales y pesqueros',
            7: 'Oficiales, operarios y artesanos de artes mecÃ¡nicas y de otros oficios',
            8: 'Operadores de instalaciones y mÃ¡quinas y ensambladores',
            9: 'Ocupaciones elementales',
            10: 'Otros no identificados',
            999: np.nan,
        }),
        
       educa=lambda x: np.select(
            [x['nivel'].notna() & x['nivel'].isin([0, 1, 2, 88, 99, 902]),
            x['nivel'].notna() & (x['nivel'] == 3),
            x['nivel'].notna() & x['nivel'].isin([4, 6]),
            x['nivel'].notna() & x['nivel'].isin([5, 7, 8]) & (x['termino_nivel'] == 1),
            x['nivel'].notna() & x['nivel'].isin([5, 7, 8]) & (x['termino_nivel'] == 2),
            x['nivel'].notna() & (x['nivel'] == 9) & (x['termino_nivel'] == 1),
            x['nivel'].notna() & (x['nivel'] == 9) & (x['termino_nivel'] == 2),
            x['nivel'].notna() & x['nivel'].isin([10, 11, 12]) & (x['termino_nivel'] == 1),
            x['nivel'].notna() & x['nivel'].isin([10, 11, 12]) & (x['termino_nivel'] == 2)],
            ["Otro nivel educativo", "EducaciÃ³n BÃ¡sica", "EducaciÃ³n Media",
            "TÃ©cnica completa", "TÃ©cnica incompleta", "Universitaria completa", 
            "Universitaria incompleta", "Postgrado", "Universitaria completa"],
            default="Otro nivel educativo"
        ),
        
        categoria_ocupacion=lambda x: x['categoria_ocupacion'].map({
            0: 'No corresponde',
            1: 'Empleador',
            2: 'Cuenta propia (independiente)',
            3: 'Asalariado sector privado',
            4: 'Asalariado sector pÃºblico',
            5: 'Personal de servicio domÃ©stico puertas afuera',
            6: 'Personal de servicio domÃ©stico puertas adentro',
        }),
        
        ocup_form=lambda x: x['ocup_form'].map({
            1: 'Ocupado formal',
            2: 'Ocupado informal',
        }),
        
        proveedor=lambda x: x['proveedor'].map({
            0: 'No corresponde',
            1: 'Proveedor principal',
        }),
        
        parentesco=lambda x: x['parentesco'].map({
            1: 'Persona proveedora principal',
            2: 'CÃ³nyuge',
            3: 'Conviviente',
            4: 'Hijo/a o hijastro/a',
            5: 'Yerno o nuera',
            6: 'Nieto/a',
        }))
        
    .rename(columns={'idrph': 'id_persona'}) # Renombrar columnas (en caso que corresponda)
    .drop_duplicates(subset=['id_persona']) # Eliminar duplicados en columnas (en caso que corresponda)
    .dropna(subset=['id_persona']) # Eliminar NaN de columnas (en caso que corresponda)
    .query('edad >= 18 and activ == 1') # Mantener rango de columnas (en caso que corresponda)
    .replace({col: ciiu4 for col in datos_proc.columns if 'rev4cl' in col}) # Reemplazar valores en columnas que contienen 'rev4cl'
)
```

```python
datos_proc = datos_proc[['id_persona', 
                         'sexo', 
                         'edad', 
                         'edad_tr', 
                         'sexo_edad', 
                         'region', 
                         'educa',
                         'habituales',
                         'efectivas',
                         'b1',
                         'activ',
                         'categoria_ocupacion',
                         'ocup_form',
                         'proveedor', 
                         'parentesco',
                         ] + 
                        [col for col in datos_proc.columns if col.startswith('c2_1')] +
                        [col for col in datos_proc.columns if 'rev4cl' in col]]

datos_proc.rename(columns={
    'c2_1_1': 'hrs_diarias',
    'c2_1_2': 'dias_sem',
    'c2_1_3': 'hrs_sem',
    'b1': 'grupoOcup_ciuo08',
    'b13_rev4cl_caenes': 'empresaPaga_ciiu4',
    'b14_rev4cl_caenes': 'empresaTrabaja_ciiu4',
    'r_p_rev4cl_caenes': 'empresaDueno_ciiu4',
}, inplace=True)
```

```python
# RevisiÃ³n de variables I
print(datos_proc.shape, "\n\n")
print(datos_proc.columns, "\n\n")
print(datos_proc.dtypes, "\n\n")
print(datos_proc['sexo'].value_counts(), '\n\n')
print(datos_proc['edad_tr'].value_counts(), '\n\n')
print(datos_proc['sexo_edad'].value_counts(), '\n\n')
print(datos_proc['region'].value_counts(), '\n\n')
print(datos_proc['educa'].value_counts(), '\n\n')
```

```python
# RevisiÃ³n de variables II
print(datos_proc['grupoOcup_ciuo08'].value_counts().sort_index(), '\n\n')
print(datos_proc['empresaPaga_ciiu4'].value_counts().sort_index(), '\n\n')
print(datos_proc['empresaTrabaja_ciiu4'].value_counts().sort_index(), '\n\n')
print(datos_proc['empresaDueno_ciiu4'].value_counts().sort_index(), '\n\n')
```
